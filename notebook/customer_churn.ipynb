# Introduction
Customer churn is defined as the rate that customers stop subscribing or using a company's services during a specific period. Customer churn occurs because of multiple reasons, including customer service, price changes, and many other factors. In this notebook, I will be investigating the causes of customer churn for telecommunications subscribers.
https://www.kaggle.com/datasets/reyhanarighy/data-telco-customer-churn/ 

## Data Exploratory

<strong>Loading the modules necessary for this project</strong>

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot  as plt

plt.rcParams["figure.figsize"] = (10, 5)
plt.rcParams["axes.labelweight"] = "bold"
plt.rcParams["axes.titleweight"] = "bold"
```

<strong>Opening the dataset we will be working with</strong>

```python
df= pd.read_csv("../raw_data/data_telco_customer_churn.csv")
df
```

<strong>Summary of the dataset</strong>

```python
df.info()
```

The dataset consists of 11 columns, which represent the potential factors influencing churn, and includes the "churn" column. It contains a total of 4,930 rows, each corresponding to individual subscribers.

```python
sns.countplot(data = df, x = 'Churn', hue='Dependents')
plt.title("Histogram of dependents who churned")
plt.ylabel('Subscribers')
plt.show()
```

```python
sns.countplot(data = df, x = 'Churn', hue='InternetService')
plt.title("Internet service subscribers who churned")
plt.ylabel('Subscribers')
plt.show()
```

From those who churned, fibre optics subscribers are the majority.

```python
sns.countplot(data = df, x = 'Churn', hue='Contract')
plt.title("subscribers on different contracts who churned")
plt.ylabel('Subscribers')
plt.show()
```

From those who churned, majority of them was on a month-to-month contract.

```python
sns.countplot(data = df, x = 'Churn', hue='TechSupport')
plt.title("subscribers with tech support who churned")
plt.ylabel('Subscribers')
plt.show()
```

From those who churned, majority of them did  not have tech support.

```python
sns.histplot(x = 'tenure', hue = 'Churn', data = df, multiple='dodge')
plt.title("Histogram of tenure and churn")
plt.ylabel('Subscribers')
plt.show()
```

Tenure is how long the customers have been subscribed to the service. The longer the customer has been with a company, the less they are to churn.
